```python
import pandas as pd 
match_info  = pd.read_csv("worldcup_predictor_teams.csv")
```

```python
from rich.console import Console
from rich.table import Table

# ---------------------------------------------
#  Get DataFrame schema and info for summary
# ---------------------------------------------

df = match_info      # use the pandas DataFrame we loaded earlier

# Count how many null (missing) values each column has
null_counts = df.isnull().sum()

# Total number of rows in the dataset (used for % null calculation)
total = len(df)

# ---------------------------------------------
#  Build a pretty summary table using Rich
# ---------------------------------------------

# Create the table and set the title shown above the table
table = Table(title="Summary of worldcup_predictor_teams.csv")

# Add the table columns (with colors and styles)
table.add_column("Name", style="cyan", no_wrap=True)         # Column name
table.add_column("Type", style="magenta")                    # Data type
table.add_column("# Nulls", style="green")                   # Number of missing values
table.add_column("% Nulls", style="yellow")                  # Percentage missing

# ---------------------------------------------
#  Loop through each column and fill the table
# ---------------------------------------------

for col in df.columns:
    dtype = df[col].dtype                     # Get the column’s data type
    n_nulls = int(null_counts[col])           # Number of missing values
    pct_null = round((n_nulls / total) * 100, 2)   # Calculate % missing

    # Add one row to the table for the current column
    table.add_row(
        col,                # Column name
        str(dtype),         # Data type
        str(n_nulls),       # # of null values
        f"{pct_null}%"      # % of null values
    )

# ---------------------------------------------
#  Display the table in the notebook
# ---------------------------------------------

console = Console()
console.print(table)
```
